```python

#each new element with "+" instead of appending as usual:
year = []

for a in progressbar.progressbar(range(10)):
    for i in progressbar.progressbar(range(100)):
        year = year + soup.select("#mw-content-text > div.mw-parser-output > div.navbox > table > tbody > tr:nth-child(" + str(a) + ") > td > div > ul > li:nth-child(" + str(i) + ") > a")
    

# check the output:
year
```

```python
import pandas as pd

song_soups = pd.DataFrame(columns=['Song', 'Artist'])
```

```python
from random import randint
from time import sleep
```

```python
for yr in year:
    # send request
    url = "https://en.wikipedia.org/" + yr["href"]
    response = requests.get(url)
    print(yr.get_text(), response.status_code)

    # parse & store html
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find('table', class_ = 'wikitable sortable')
    # Collecting Ddata
    for row in table.tbody.find_all('tr'):    
        # Find all data for each column
        columns = row.find_all('td')

        if(len(columns)>2):
            song = columns[1].text.strip('"')
            artist = columns[2].text.strip()

            song_soups=song_soups.append({'Song': song,  'Artist': artist}, ignore_index=True)

    # respectful nap:
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```
